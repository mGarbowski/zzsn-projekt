```python
from models.diffusion import WrappedDiffusion, GenerationParams
import matplotlib.pyplot as plt
```

```python
diffusion = WrappedDiffusion.from_pretrained(artifact_id)
```

```python
styles = list(multipliers.keys())
```

```python
def make_with_intervention(diffusion, style, multipliers):
    params = GenerationParams(
        prompts=[f"A british shorthair cat in {style} style"],
        num_seeds=1,
        num_inference_steps=50,
        guidance_scale=7.5
    )
    return diffusion.generate_with_intervention(params, intervention_multipliers=multipliers[style])[0].image
```

```python
images = [
    make_with_intervention(diffusion, style, multipliers)
    for style in styles
]

fix, axes = plt.subplots(len(images))
for ax in zip(axes, images, styles):
    ax[0].imshow(ax[1])
    ax[0].set_title(ax[2])
    ax[0].axis("off")
plt.tight_layout()
plt.show()
```
