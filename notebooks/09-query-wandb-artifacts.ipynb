# Pobranie nazw artefaktów przez API wandb

```python
import re
import wandb

api = wandb.Api()

entity = "mikolaj-garbowski-warsaw-university-of-technology"
project = "zzsn-projekt"
artifact_type = "model"

pattern = re.compile(r"^model-.*-step_.*$")

names = []
for c in api.artifact_type(type_name=artifact_type, project=f"{entity}/{project}").collections():
    base = c.name.split("/")[-1]
    if pattern.match(base):
        names.append(base)

names = sorted(set(names))
print(names)
```

```python
len(names)
```

```python
ids = [
    f"{entity}/{project}/{name}:v0"
    for name in names
]
ids
```

```python
ids_arg = ",".join(ids)
ids_arg
```
